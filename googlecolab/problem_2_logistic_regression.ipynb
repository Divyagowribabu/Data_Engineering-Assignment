PROBLEM 2: Credit Card Fraud Detection
(Logistic Regression)

A bank wants to detect fraudulent transactions in real time.
Each transaction must be classified as:

0 → Legitimate  

1 → Fraud

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/credit_card_transactions.csv')

X = df.drop(['TransactionID', 'IsFraud'], axis=1)
y = df['IsFraud']
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
y_probs = model.predict_proba(X)[:,1]

fpr, tpr, _ = roc_curve(y, y_probs)

plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```

```python
model_l1 = LogisticRegression(penalty='l1', solver='liblinear')
model_l2 = LogisticRegression(penalty='l2')

model_l1.fit(X, y)
model_l2.fit(X, y)
```
